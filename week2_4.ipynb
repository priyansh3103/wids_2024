```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target function
def objective_function(x):
    return (x[0] - 1)**2 + x[1]**2 + (x[2] + 1)**2

# Define the gradient of the function
def compute_gradient(x):
    grad = np.zeros((3, 1))
    grad[0][0] = 2 * (x[0] - 1)
    grad[1][0] = 2 * x[1]
    grad[2][0] = 2 * (x[2] + 1)
    return grad

# Compute the norm of the gradient
def calculate_norm(grad):
    return np.linalg.norm(grad)

# Proximal gradient descent implementation
def proximal_gradient_descent(initial_x, step_size, reg_param, max_iterations):
    x = initial_x
    iteration = 0

    while iteration < max_iterations:
        grad = compute_gradient(x)
        intermediate_step = x - step_size * grad

        # Check if the solution remains within the feasible region
        for i in range(len(intermediate_step)):
            if intermediate_step[i] < 0:
                intermediate_step[i] = 0  # Project onto the feasible set

        x = intermediate_step
        iteration += 1

    # Compute the final function value and error
    final_value = objective_function(x)
    error = final_value  # Since the optimal value is 0
    return x, final_value, error

# Parameters
initial_x = np.array([[0], [0], [0]])
step_size = 0.3  # Step size
regularization_param = 0.5
iteration_values = [100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000, 5000000]

# Arrays to store results for plotting
iterations = []
errors = []

# Run the proximal gradient algorithm for different iteration limits
for max_iter in iteration_values:
    minimizer, minimum_value, error_value = proximal_gradient_descent(
        initial_x, step_size, regularization_param, max_iter
    )

    print(f"Max Iterations: {max_iter}")
    print(f"Minimizer: {minimizer}")
    print(f"Minimum Value: {minimum_value}")
    print(f"Error: {error_value}\n")

    iterations.append(max_iter)
    errors.append(error_value * 100)  # Scale error for visualization

# Plot error vs. iterations
plt.plot(iterations, errors, marker="o")
plt.xlabel("Number of Iterations")
plt.ylabel("Error (%)")
plt.title("Error vs. Iterations for Proximal Gradient Descent")
plt.grid(True)
plt.show()
```
