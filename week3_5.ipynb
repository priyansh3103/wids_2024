```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loss functions
def G_h(z):
    return np.maximum(0, 1 - z)

def G_l(z):
    return np.log(1 + np.exp(-z))

def G_sh(z):
    return np.maximum(0, 1 - z)**2

# Define the range of z values
z = np.linspace(-3, 3, 500)

# Compute the loss values
hinge_loss = G_h(z)
logistic_loss = G_l(z)
squared_hinge_loss = G_sh(z)

# Plot the loss functions
plt.figure(figsize=(10, 6))
plt.plot(z, hinge_loss, label="Hinge Loss (Gh(z))", color="red")
plt.plot(z, logistic_loss, label="Logistic Loss (Gl(z))", color="blue")
plt.plot(z, squared_hinge_loss, label="Squared Hinge Loss (Gsh(z))", color="green")

# Add labels, legend, and grid
plt.title("Loss Functions for Binary Classification", fontsize=14)
plt.xlabel("z", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.axhline(0, color="black", linestyle="--", linewidth=0.7)
plt.axvline(0, color="black", linestyle="--", linewidth=0.7)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)

# Display the plot
plt.show()
```
